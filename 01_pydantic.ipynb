```python
from pydantic import BaseModel
```

```python
class Student(BaseModel):
    name:str
    grade:str
    age:int
    city:str
```

```python
s1 = Student(name="John", grade="X", age=15, city="New York")
s1
```

```python
s1.model_dump()
```

```python
s1 = Student(name="John", grade="X", age=15, city=10)
```

```python
from typing import Optional, List

class Book(BaseModel):
    title:str
    author:str
    num_pages:int
    category:Optional[str] = "General"
    important_pages: List[int] = None
```

```python
b1 = Book(title="HP", author="JKR", num_pages=500)
b1
```
